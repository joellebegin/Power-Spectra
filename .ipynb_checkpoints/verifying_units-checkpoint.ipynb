```python
import numpy as np 
import matplotlib.pyplot as plt
from powerSpectrum import p_spec
```

```python
def plane_sinusoid(n, lambd, xmax):
    x = np.linspace(0, xmax, n)
    data = np.sin(x*(2*np.pi/lambd))
    return np.array([data for i in range(n)])

graph = plane_sinusoid(1000, 20, 80)
```

in order to verify that the units are working properly, I constructed a configuration space box consisting of a sinusoid. This is a good way to verify the units, since we would expect the power spectrum to peak at $k_{peak} = \frac{2\pi}{\lambda}$, $\lambda$ being the wavelength in configuration space. 

setting a pixel resolution of 5 pc, we expect $\lambda = 1250$ pc given the way we've constructed the plane wave; this is shown in the figure below. 

```python
fig, ax = plt.subplots(figsize = (20,10))
im = ax.imshow(graph, cmap=plt.cm.GnBu, origin= 'lower')

colorbar_ax = fig.add_axes()
fig.colorbar(im, cax=colorbar_ax)

ax.vlines(x = 63, ymin = 0, ymax =999)
ax.vlines(x = 313, ymin = 0, ymax =999)

ax.set_xticks([63, 200, 313, 500, 800])
ax.set_xticklabels([315, 1000, 1565, 2500, 4000])
ax.set_xlabel('Length (pc)')

ax.set_yticks([200, 500, 800])
ax.set_yticklabels([1000, 2500, 4000])
ax.set_ylabel('Length (pc)')
```
